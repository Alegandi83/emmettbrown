```python
%sql
CREATE OR REPLACE VIEW dad_open_data.locations.world AS
select      -- cities attributes
            c.id                as cities_id,
            c.name              as cities_name,
            c.latitude          as cities_latitude,
            c.longitude         as cities_longitude,
            c.native            as cities_native,
            c.timezone          as cities_timezone,
            c.translations      as cities_translations,
            c.wikiDataId        as cities_wikiDataId,
            -- h3 polygon ids for each resolution (0-15) for cities
            h3_longlatash3(c.longitude, c.latitude, 0)  as cities_h3_res_0,
            h3_longlatash3(c.longitude, c.latitude, 1)  as cities_h3_res_1,
            h3_longlatash3(c.longitude, c.latitude, 2)  as cities_h3_res_2,
            h3_longlatash3(c.longitude, c.latitude, 3)  as cities_h3_res_3,
            h3_longlatash3(c.longitude, c.latitude, 4)  as cities_h3_res_4,
            h3_longlatash3(c.longitude, c.latitude, 5)  as cities_h3_res_5,
            h3_longlatash3(c.longitude, c.latitude, 6)  as cities_h3_res_6,
            h3_longlatash3(c.longitude, c.latitude, 7)  as cities_h3_res_7,
            h3_longlatash3(c.longitude, c.latitude, 8)  as cities_h3_res_8,
            h3_longlatash3(c.longitude, c.latitude, 9)  as cities_h3_res_9,
            h3_longlatash3(c.longitude, c.latitude, 10) as cities_h3_res_10,
            h3_longlatash3(c.longitude, c.latitude, 11) as cities_h3_res_11,
            h3_longlatash3(c.longitude, c.latitude, 12) as cities_h3_res_12,
            h3_longlatash3(c.longitude, c.latitude, 13) as cities_h3_res_13,
            h3_longlatash3(c.longitude, c.latitude, 14) as cities_h3_res_14,
            h3_longlatash3(c.longitude, c.latitude, 15) as cities_h3_res_15,
            -- states attributes
            s.id                as state_id,
            s.name              as state_name,
            s.fips_code         as state_fips_code,
            s.iso2              as state_iso2,
            s.iso3166_2         as state_iso3166_2,
            s.type              as state_type,
            s.level             as state_level,
            s.parent_id         as state_parent_id,
            s.native            as state_native,
            s.latitude          as state_latitude,
            s.longitude         as state_longitude,
            s.timezone          as state_timezone,
            s.translations      as state_translations,
            s.wikiDataId        as state_wikiDataId, 
            -- h3 polygon ids for each resolution (0-15) for states
            h3_longlatash3(s.longitude, s.latitude, 0)  as state_h3_res_0,
            h3_longlatash3(s.longitude, s.latitude, 1)  as state_h3_res_1,
            h3_longlatash3(s.longitude, s.latitude, 2)  as state_h3_res_2,
            h3_longlatash3(s.longitude, s.latitude, 3)  as state_h3_res_3,
            h3_longlatash3(s.longitude, s.latitude, 4)  as state_h3_res_4,
            h3_longlatash3(s.longitude, s.latitude, 5)  as state_h3_res_5,
            h3_longlatash3(s.longitude, s.latitude, 6)  as state_h3_res_6,
            h3_longlatash3(s.longitude, s.latitude, 7)  as state_h3_res_7,
            h3_longlatash3(s.longitude, s.latitude, 8)  as state_h3_res_8,
            h3_longlatash3(s.longitude, s.latitude, 9)  as state_h3_res_9,
            h3_longlatash3(s.longitude, s.latitude, 10) as state_h3_res_10,
            h3_longlatash3(s.longitude, s.latitude, 11) as state_h3_res_11,
            h3_longlatash3(s.longitude, s.latitude, 12) as state_h3_res_12,
            h3_longlatash3(s.longitude, s.latitude, 13) as state_h3_res_13,
            h3_longlatash3(s.longitude, s.latitude, 14) as state_h3_res_14,
            h3_longlatash3(s.longitude, s.latitude, 15) as state_h3_res_15,
            -- countries attributes
            cn.id               as country_id,
            cn.name             as country_name,
            cn.iso2             as country_iso2,
            cn.iso3             as country_iso3,
            cn.numeric_code     as country_numeric_code,
            cn.phonecode        as country_phonecode,
            cn.capital          as country_capital,
            cn.currency         as country_currency,
            cn.currency_name    as country_currency_name,
            cn.currency_symbol  as country_currency_symbol,
            cn.tld              as country_tld,
            cn.native           as country_native,
            cn.population       as country_population,
            cn.gdp              as country_gdp,
            cn.nationality      as country_nationality,
            cn.timezones        as country_timezones,
            cn.translations     as country_translations,
            cn.latitude         as country_latitude,
            cn.longitude        as country_longitude,
            cn.emoji            as country_emoji,
            cn.wikiDataId       as country_wikiDataId,
            -- h3 polygon ids for each resolution (0-15) for countries
            h3_longlatash3(cn.longitude, cn.latitude, 0)  as country_h3_res_0,
            h3_longlatash3(cn.longitude, cn.latitude, 1)  as country_h3_res_1,
            h3_longlatash3(cn.longitude, cn.latitude, 2)  as country_h3_res_2,
            h3_longlatash3(cn.longitude, cn.latitude, 3)  as country_h3_res_3,
            h3_longlatash3(cn.longitude, cn.latitude, 4)  as country_h3_res_4,
            h3_longlatash3(cn.longitude, cn.latitude, 5)  as country_h3_res_5,
            h3_longlatash3(cn.longitude, cn.latitude, 6)  as country_h3_res_6,
            h3_longlatash3(cn.longitude, cn.latitude, 7)  as country_h3_res_7,
            h3_longlatash3(cn.longitude, cn.latitude, 8)  as country_h3_res_8,
            h3_longlatash3(cn.longitude, cn.latitude, 9)  as country_h3_res_9,
            h3_longlatash3(cn.longitude, cn.latitude, 10) as country_h3_res_10,
            h3_longlatash3(cn.longitude, cn.latitude, 11) as country_h3_res_11,
            h3_longlatash3(cn.longitude, cn.latitude, 12) as country_h3_res_12,
            h3_longlatash3(cn.longitude, cn.latitude, 13) as country_h3_res_13,
            h3_longlatash3(cn.longitude, cn.latitude, 14) as country_h3_res_14,
            h3_longlatash3(cn.longitude, cn.latitude, 15) as country_h3_res_15,
            -- subregions attributes
            sr.id               as subregion_id,
            sr.name             as subregion_name,
            sr.translations     as subregion_translations,
            sr.wikiDataId       as subregion_wikiDataId,
            -- regions attributes
            r.id                as region_id,
            r.name              as region_name,
            r.translations      as region_translations,
            r.wikiDataId        as region_wikiDataId
from        das_geo_psql_db.public.cities as c
inner join  das_geo_psql_db.public.states as s
  on        c.state_id = s.id
inner join  das_geo_psql_db.public.countries as cn
  on        s.country_id = cn.id
inner join  das_geo_psql_db.public.subregions as sr
  on        cn.subregion_id = sr.id
inner join  das_geo_psql_db.public.regions as r
  on        sr.region_id = r.id
```
